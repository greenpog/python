```python
import pandas as pd

# Загрузка датасета
df = pd.read_csv("students.csv")
# Первые 5 строк
print("Первые 5 строк:")
print(df.head())
# Информация о данных
print("\nИнформация о данных:")
print(df.info())
# Статистика по числовым значениям
print("\nСтатистика по числовым столбцам:")
print(df.describe())
# Средний балл студентов
average_score = df['Score'].mean()
print(f"\nСредний балл студентов: {average_score:.2f}")
# Количество студентов в каждой группе
group_counts = df['Group'].value_counts()
print("\nКоличество студентов в каждой группе:")
print(group_counts)
```

```python
import pandas as pd

# Загрузка данных
df = pd.read_csv("students.csv")
# Фильтрация: студенты с баллом выше 80
high_scores = df[df['Score'] > 80]
print("Студенты с баллом выше 80:")
print(high_scores)
# Сортировка по убыванию балла
sorted_high_scores = high_scores.sort_values(by='Score', ascending=False)
print("\nСтуденты с баллом выше 80 (отсортированы по убыванию балла):")
print(sorted_high_scores)
# Самый старший студент
oldest_student = df[df['Age'] == df['Age'].max()]
print("\nСамый старший студент:")
print(oldest_student)
# Самый младший студент
youngest_student = df[df['Age'] == df['Age'].min()]
print("\nСамый младший студент:")
print(youngest_student)
```

```python
import pandas as pd

# Загрузка данных
df = pd.read_csv("students.csv")

# Проверка наличия пропущенных значений
print("Количество пропущенных значений по столбцам:")
print(df.isnull().sum())

# Заполнение пропусков в столбце Score средним значением
if df['Score'].isnull().any():
    mean_score = df['Score'].mean()
    df['Score'].fillna(mean_score, inplace=True)

# Удаление строк с пропусками в столбце Group
df.dropna(subset=['Group'], inplace=True)

# Проверка после обработки
print("\nПроверка данных после обработки пропусков:")
print(df.isnull().sum())
```

```python
import pandas as pd

# Загрузка данных
df = pd.read_csv("students.csv")
# Добавление столбца Passed
df['Passed'] = df['Score'].apply(lambda x: 1 if x >= 60 else 0)
# Группировка по столбцу Group с агрегацией
group_stats = df.groupby('Group').agg({
    'Score': 'mean',       # средний балл
    'Age': 'median'        # медианный возраст
}).rename(columns={'Score': 'Average_Score', 'Age': 'Median_Age'})
# Вывод результатов
print("Агрегация по группам:")
print(group_stats)

print("\nТаблица с новым столбцом Passed:")
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv("students.csv")
# Гистограмма распределения баллов
plt.figure(figsize=(10, 5))
plt.hist(df['Score'], bins=10, color='purple', edgecolor='black')
plt.title('Распределение баллов студентов')
plt.xlabel('Балл')
plt.ylabel('Количество студентов')
plt.tight_layout()
plt.show()
# Столбчатая диаграмма среднего балла по группам
group_avg_scores = df.groupby('Group')['Score'].mean().sort_index()

plt.figure(figsize=(8, 5))
plt.bar(group_avg_scores.index, group_avg_scores.values, color='purple', edgecolor='black')
plt.title('Средний балл по группам')
plt.xlabel('Группа')
plt.ylabel('Средний балл')
plt.tight_layout()
plt.show()
```
